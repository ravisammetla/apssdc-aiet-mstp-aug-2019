### Question -1:
write a program to print the digits in given string using map or filter

- Input Format: 1b3a85a05H4
- Output Format: 1385054

```python
n=input()
def isDigit(c):
    c=str(c)
    if c.isdigit():
        return True
    return False
''.join(list(filter(isDigit,n)))
```

### Question -2:
write a program to print the N qubes sum from 1 to N if n=3(13+23+3**3)

- Input Format: 3
- Output Format: 36

```python
result=lambda x:x**3
sum=0
for i in range(1,int(input())+1):
    sum+=result(i)
print(sum)
```
